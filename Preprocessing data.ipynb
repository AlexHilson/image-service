### Processing

The levels need converting from height to altitude. Fortunately, I just spoke to Phil, and AVD already have code to do this! It requres Cython so needs Monty, compiled using

    python2.7 setup.py install --single-version-externally-managed --record record.txt
    
IMPORTANT: this regridding is proprietry and we will need PERMISSION from Stuart Bell to release the 500 lines of restratification code.

Other things Phil said:

1. We can get round the linear interpolation of log levels by creating an aux coord of log(height) and doing interpolation based on that
2. We can downscale data with interpolation/regridding using the area weighted stuff.
3. If we load in the topography field at the same time (and the are both pp files) then it calculates the altitude for us)

```python
import iris
data, _ = iris.load(["/Users/niall/Data/ukv/synth_cloud_frac.pp", "/Users/niall/Data/ukv/ukv_orog.pp"])
data = data[:3]
```

We've now loaded the data, and because we loaded in the topography at the same time, we now have a derived coord of height. We can now use the Monty `vinterp` to restratify the data. This is also the stage where we might want to have a log space derived coord.

```python
sea_level = 4.5
sanitized_alt = data.coord("altitude").points
sanitized_alt[np.logical_not(np.isfinite(sanitized_alt))] = sea_level
sanitized_alt[sanitized_alt < sea_level] = sea_level
logaltcoord = iris.coords.AuxCoord(np.log(sanitized_alt), long_name="log_altitude")
data.add_aux_coord(logaltcoord, [1, 2, 3])
```

```python
import sys
sys.path.append("/Users/niall/Projects/monty/lib/")
import monty.vinterp
```

```python
nlevels = 35
log_levs = np.linspace(data.coord("log_altitude").points.min(),
                                                  data.coord("log_altitude").points.max(),
                                                  nlevels)
restratified_data = monty.vinterp.interpolate(log_levs,
                                                data.coord("log_altitude").points,
                                                data.data,
                                                axis=1,
                                                extrapolation=monty.vinterp.EXTRAPOLATE_NAN,
                                                interpolation=monty.vinterp.INTERPOLATE_LINEAR)
```

```python
restratified_data_cube = data[:, :nlevels, :, :].copy(data=np.ma.masked_invalid(restratified_data))
restratified_data_cube.remove_coord("model_level_number")
restratified_data_cube.remove_coord("sigma")
restratified_data_cube.remove_coord("level_height")
restratified_data_cube.remove_aux_factory(restratified_data_cube.aux_factories[0])

restratified_data_cube.coord("grid_latitude").guess_bounds()
restratified_data_cube.coord("grid_longitude").guess_bounds()

alt_coord = iris.coords.DimCoord(np.exp(log_levs), long_name="altitude", units="m")
restratified_data_cube.add_dim_coord(alt_coord, 1)
```

### Now to regrid to a rectalinear grid and downscale

We have now dealt with the log levels and the topography. The next task is the regrid the data from a polar stereographic projection into something that is rectalinear. We will also deal with any downscaling that we are doing at the same time.

```python
rl = iris.load_cube("/Users/niall/Data/ukv/uk_low_cloud.nc")
min_lat = rl.coord('latitude').points.min()
max_lat = rl.coord('latitude').points.max()
min_lon = rl.coord('longitude').points.min()
max_lon = rl.coord('longitude').points.max()

u = iris.unit.Unit("degrees")
cs = iris.coord_systems.GeogCS(iris.fileformats.pp.EARTH_RADIUS)

nlat = 400
nlon = 400
latc = iris.coords.DimCoord(np.linspace(min_lat, max_lat, nlat), standard_name="latitude", units=u, coord_system=cs)
latc.guess_bounds()
lonc = iris.coords.DimCoord(np.linspace(min_lon, max_lon, nlon), standard_name="longitude", units=u, coord_system=cs)
lonc.guess_bounds()

grid_cube = iris.cube.Cube(np.empty([nlat, nlon]))
grid_cube.add_dim_coord(latc, 0)
grid_cube.add_dim_coord(lonc, 1)
```

```python
rg_data_cube = restratified_data_cube.regrid(grid_cube, iris.analysis.Linear(extrapolation_mode='mask'))
```

```python
def getPostDict(cube, mime_type="image/png"):
    """
    Converts relevant cube metadata into a dictionary of metadata which is compatable
    with the data service.

    """
    with iris.FUTURE.context(cell_datetime_objects=True):
        payload = {'forecast_reference_time': cube.coord("forecast_reference_time").cell(0).point.isoformat()+".000Z",
                   'forecast_time' : cube.coord("time").cell(0).point.isoformat()+".000Z",
                   'phenomenon' : cube.var_name,
                   'mime_type' : mime_type,
                   'model' : 'uk_v'}#,
                   # 'data_dimensions': {'x': cube.shape[0], 'y': cube.shape[1], 'z': cube.shape[2]}}
        return payload
```
